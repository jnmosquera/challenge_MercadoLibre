```python
import requests
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
ACCESS_TOKEN = "APP_USR-xxxxx" #Para ejecutar los GET a la API, insertar ACCESS_TOKEN personal
headers = {
    "Authorization": f"Bearer {ACCESS_TOKEN}"
}
```

```python
# Uso el Api para traer los resultados de 3 busquedas:

# Nintendo Switch
df_nintendo = pd.DataFrame(requests.get("https://api.mercadolibre.com/sites/MLA/search?q=Nintendo Switch&limit=50#json", headers=headers ).json()['results'])

#PlayStation5
df_playstation = pd.DataFrame(requests.get("https://api.mercadolibre.com/sites/MLA/search?q=PlayStation 5&limit=50#json", headers=headers).json()['results'])

#XBox Serie X
df_xbox = pd.DataFrame(requests.get("https://api.mercadolibre.com/sites/MLA/search?q=XBox Serie x&limit=50#json", headers=headers).json()['results'])
```

```python
# Uno todos los dataframes obtenidos
dataframes = [df_nintendo, df_playstation, df_xbox]

all_consoles_df = pd.concat(dataframes, ignore_index=True)
```

```python
all_consoles_df.head()
```

```python
#Recorro el listado de items_id y por cada item_id mando el GET. Al final uno todos los resultados.
for indx, i in enumerate(all_consoles_df['id']):
    if indx==0:
        df= pd.DataFrame(requests.get("https://api.mercadolibre.com/items?ids="+i,headers=headers).json())

    else:
        df_aux= pd.DataFrame(requests.get("https://api.mercadolibre.com/items?ids="+i,headers=headers).json())
        df = pd.concat([df,df_aux],axis=0)
```

```python
#Normalizo el resultado y selecciono las columnas que considero relevante un futuro analisis

all_items_df = pd.json_normalize(df['body'])
all_items_df.head()

selected_columns = [
    'id', 'title', 'category_id', 'official_store_id','price', 'base_price', 'original_price', 
    'currency_id', 'initial_quantity', 'condition', 'international_delivery_mode', 'status', 'warranty', 
    'shipping.logistic_type', 'shipping.store_pick_up', 'seller_address.city.name'
]

consoles_filtered_df = all_items_df[selected_columns].copy()
```
